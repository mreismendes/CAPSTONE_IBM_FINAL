```
#@title IMPORTING DATA
%load_ext google.colab.data_table
import pandas as pd
import numpy as np
df = pd.read_excel('/content/drive/My Drive/Consultoria/L3/MARGEM_CONTRIBUICAO/BD_CENARIOS.xlsx',sheet_name="BD",index_col='INDEX')
```

```
#@title Parâmetros Margem de Contribuição

Peso_Compra = "210" #@param [180,210]
Ágio_Resposição = "0.25000000000000006" #@param [0.1,0.15000000000000002,0.30000000000000004,0.25000000000000006,0.20000000000000004,0.3500000000000001,0.40000000000000013]
Rendimento_Carcaça = "0.54" #@param [0.53, 0.54, 0.56, 0.55, 0.5700000000000001]

cond1 = df['PESO_COMPRA']==float(Peso_Compra)
cond2 = df['AGIO_REPOSICAO']==float(Ágio_Resposição)
cond3 = df['RENDIMENTO_CARCACA']==float(Rendimento_Carcaça)
df_plot = df[cond1&cond2&cond3]

max_margem = df_plot.MARGEM.max()
min_margem = df_plot.MARGEM.min()

max_boi = df_plot.PRECO_ARROBA_BOI.max()
min_boi = df_plot.PRECO_ARROBA_BOI.min()

max_milho = df_plot.PRECO_MILHO.max()
min_milho = df_plot.PRECO_MILHO.min()

max_mtir = df_plot.MTIR_ANO.max()
min_mtir = df_plot.MTIR_ANO.min()
################
import plotly.graph_objects as go
import numpy as np

max_margem = df_plot.MARGEM.max()
min_margem = df_plot.MARGEM.min()

max_boi = df_plot.PRECO_ARROBA_BOI.max()
min_boi = df_plot.PRECO_ARROBA_BOI.min()

max_milho = df_plot.PRECO_MILHO.max()
min_milho = df_plot.PRECO_MILHO.min()

max_mtir = df_plot.MTIR_ANO.max()
min_mtir = df_plot.MTIR_ANO.min()
################
import plotly.graph_objects as go
import numpy as np

fig1 = go.Figure(data=[go.Mesh3d(x=(df_plot.PRECO_ARROBA_BOI),
                   y=(df_plot.PRECO_MILHO),
                   z=(df_plot.MARGEM),
                   opacity=0.5,
                   color='darkslategray'
                  )])

fig1.update_layout(
    scene = dict(
        xaxis = dict(nticks=20, range=[min_boi,max_boi],),
                     yaxis = dict(nticks=10, range=[min_milho,max_milho],),
                     zaxis = dict(nticks=10, range=[min_margem,max_margem],),),
    width=700,
    margin=dict(r=20, l=10, b=10, t=10))

fig1.update_layout(scene = dict(
                    xaxis_title='PREÇO BOI R$/@',
                    yaxis_title='PREÇO MILHO R$/SC',
                    zaxis_title='MARGEM DE CONTRIBUIÇÃO'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))



#################

fig2 = go.Figure(data=[go.Mesh3d(x=(df_plot.PRECO_ARROBA_BOI),
                   y=(df_plot.PRECO_MILHO),
                   z=(df_plot.MTIR_ANO),
                   opacity=0.5,
                   color='darkslategray'
                  )])

fig2.update_layout(
    scene = dict(
        xaxis = dict(nticks=20, range=[min_boi,max_boi],),
                     yaxis = dict(nticks=10, range=[min_milho,max_milho],),
                     zaxis = dict(nticks=10, range=[min_mtir,max_mtir],),),
    width=700,
    margin=dict(r=20, l=10, b=10, t=10))

fig2.update_layout(scene = dict(
                    xaxis_title='PREÇO BOI R$/@',
                    yaxis_title='PREÇO MILHO R$/SC',
                    zaxis_title='TIR a.a'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))






fig1.show()
fig2.show()
```
